```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris = load_iris()
dir(iris)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
lr_scores = cross_val_score(LogisticRegression(), iris.data, iris.target)
np.average(lr_scores)
```

```python
svc_scores = cross_val_score(SVC(), iris.data, iris.target)
np.average(svc_scores)
```

```python
dt_scores = cross_val_score(DecisionTreeClassifier(), iris.data, iris.target)
np.average(dt_scores)
```

```python
rfc_scores = cross_val_score(RandomForestClassifier(n_estimators=50), iris.data, iris.target)
np.average(rfc_scores)
```

My best evaluated score is from LogisticRegression :)
