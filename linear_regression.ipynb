```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (USD)')
plt.scatter(df.per_capita_income_usd, df.year)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income_usd)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (USD)')
plt.scatter(df.year, df.per_capita_income_usd, color = 'red')
plt.plot(df.year, reg.predict(df[['year']]))
```
