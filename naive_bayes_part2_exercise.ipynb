```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
wine = load_wine()
# dir(wine)
wine.target
```

```python
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df.head()
```

```python
df_target = pd.Series(wine.target)
df['target'] = df_target
df.tail()
```

```python
inputs = df.drop('target', axis='columns')
target = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test[:10]
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model_ml = MultinomialNB()
model_ml.fit(X_train, y_train)
```

```python
model_ml.predict(X_test)
```

```python
y_test[:10]
```

```python
model_ml.score(X_test, y_test)
```
